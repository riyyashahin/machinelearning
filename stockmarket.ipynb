```python
import numpy as np 
def create_sequences(X, y, seq_length=20):
    X_seq, y_seq = [], []
    for i in range(len(X) - seq_length):
        X_seq.append(X[i:i+seq_length])  # Take past `seq_length` days
        y_seq.append(y[i+seq_length])  # Target is the next day's price
    return np.array(X_seq), np.array(y_seq)

```

```python
from tensorflow.keras.optimizers import Adam

model.compile(optimizer=Adam(learning_rate=0.0005), loss='mean_squared_error', metrics=['mae'])
```
